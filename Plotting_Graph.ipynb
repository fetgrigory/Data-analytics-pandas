```python
# Load data from 'data_4.csv' with UTF-8 encoding
df = pd.read_csv('data\data_4.csv', encoding='utf-8')
df
```

```python
# Counting the number of items sold by type
product_counts = df['Product'].value_counts()
print(product_counts)
```

```python
 # Unique product names
product_names = product_counts.index
 # Number of units sold for each product
quantities = product_counts.values
```

```python
# Set the chart size
plt.figure(figsize=(10, 6))

# Create a bar chart with product names and their quantities
# Specify the bar color (skyblue) and edge color (black)
plt.bar(product_names, quantities, color='skyblue', edgecolor='black')

# Add a label for the X-axis (products)
plt.xlabel('Product', fontsize=12)

# Add a label for the Y-axis (quantity sold)
plt.ylabel('Quantity Sold', fontsize=12)

# Add a title for the chart
plt.title('Quantity of Products Sold by Type', fontsize=14)

# Rotate the X-axis captions and (horizontal alignment) aligns the text to the right.
plt.xticks(rotation=45, ha='right')

# This command automatically adjusts the sizes of all elements (for example, captions and headings) so that they do not overlap and do not go beyond the boundaries of the graph.
plt.tight_layout()

# Display the bar chart
plt.show()
```
