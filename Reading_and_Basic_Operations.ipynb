```python
# Load data from 'file.csv' with UTF-8 encoding
df = pd.read_csv('file.csv', encoding='utf-8')
df
```

```python
df.columns = df.columns.str.strip() # Removing extra spaces from column headings
```

```python
df.dropna(inplace=True) # Remove empty values (missing values, NA)
```

```python
df.drop(columns=['Order ID'], inplace=True) # Remove the 'Order ID' column from the DataFrame 
```

```python
# Rename columns for easier access
df.rename(columns={
   'Quantity Ordered' : 'Quantity',
   'Price Each' : 'Price',
   'Order Date' : 'Date',
   'Purchase Address': 'Address'
}, inplace=True)
df
```

```python
df.info() # dataframe information
```

```python
df['Date'] = pd.to_datetime(df.Date, format='%m/%d/%y %H:%M') # Convert 'Date' column to datetime format
df
```

```python
df.info() # dataframe information
```

```python
df['Total'] = df['Quantity'] * df ['Price'] # Calculate total price and store in 'Total' column
df
```

```python
df.to_csv('file-2.csv', encoding='utf-8', index=False) # Save DataFrame to 'file-2.csv' with UTF-8 encoding
df
```
