```python
# Load data from 'file-2.csv' with UTF-8 encoding
df = pd.read_csv('file-2.csv', encoding='utf-8')
df
```

```python
# Get the total number of records (rows) in the DataFrame df
df.shape[0]
```

```python
# Initialize a counter variable to number the output CSV files
num = 1
# Loop through the DataFrame in chunks of 5000 rows
for i in range(0, df.shape[0], 5000):
# Select a subset of 5000 rows starting from the current index
    tmp = df.loc[i:i+5000-1]
# Save the subset to a CSV file with a numbered filename
    tmp.to_csv(f'data/data_{num}.csv', index=False)
# Increment the counter for the next file
    num = num + 1
```

```python
df.loc[0:5000] # Select rows from index 0 to 5000 (inclusive) from the DataFrame
```

```python
# Initialize an empty DataFrame to store the combined data
df_all = pd.DataFrame()
# Loop through all files in the 'data/' directory
for f in os.listdir('data/'):
# Read the current CSV file into a temporary DataFrame 
    tmp = pd.read_csv('data/' + f)
# Append the data from the current file to the main DataFrame
    df_all = pd.concat([df_all, tmp])
# Display the combined DataFrame
df_all
```

```python
df_all.reset_index(inplace=True, drop=True)  # Reset the index of the DataFrame and drop the old index
# Display the combined DataFrame
df_all
```
